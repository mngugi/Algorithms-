```python
import tkinter as tk
import fibbo_lib

window = tk.Tk()
window.title("Fibonacci")
window.geometry("400x300")

def show_fibbo():
    fib_text = fibbo_lib.fibbo_function()
    output.delete("1.0", tk.END) #insert at the end 
    output.insert("1.0", fib_text)


btn = tk.Button(window, text="Show Fibonacci", command=show_fibbo)
btn.pack(pady=10)

output = tk.Text(window)
output.pack(fill=tk.BOTH, expand=True)

window.mainloop()
```
