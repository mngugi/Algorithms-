```python
arr = [0,2,6,1]

if arr[0] > arr[1]:
    arr[0], arr[1] = arr[1] , arr[0]

if arr[1] > arr[2]:
    arr[1], arr[2] = arr[2], arr[1]

if arr[2] > arr[3]:
    arr[2], arr[3] = arr[3], arr[2]


print(arr)
```

- Compare 0 and 2 → no swap
- Compare 2 and 6 → no swap
- Compare 6 and 1 → swap
- Array becomes: [0, 2, 1, 6]


```python
for i in range(len(arr) - 1):
    if arr[i] > arr[i+1]:
        arr[i], arr[i+1]  = arr[i+i], arr[i]

print(arr)        
    
```

```python
for _ in range(len(arr)):
    for i in range(len(arr) - 1):
        if arr[i] > arr[i+1]:
            arr[i], arr[i+1] = arr[i+1], arr[i]

print(arr)            
```
