```python
new_array = [0.1, 0.61, 0.9, 0.199, 0.899]
new_array
```

```python
print(new_array[2])
```

```python
#Create a variable 'minVal' and set it equal to the first value of the array.
min_val = new_array[0] 
```

```python
# Go through every element in the array.
 
for i in new_array:
    if i < min_val: # element is the array is less swap 
        min_val = i

print(f'Lowest element value in the array is {min_val}')        
    
    
```
