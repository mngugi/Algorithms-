```python
arr = [0,2,6,1]

if arr[0] > arr[1]:
    arr[0], arr[1] = arr[1] , arr[0]

if arr[1] > arr[2]:
    arr[1], arr[2] = arr[2], arr[1]

if arr[2] > arr[3]:
    arr[2], arr[3] = arr[3], arr[2]


print(arr)
```

```python
for i in range(len(arr) - 1):
    if arr[i] > arr[i+1]:
        arr[i], arr[i+1]  = arr[i+i], arr[i]

print(arr)        
    
```
