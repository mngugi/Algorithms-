```python
bubble_selection_array = [64, 34, 25, 12, 22,9,21, 11, 90, 5]

n = len(bubble_selection_array)
for i in range(n):
    min_index = i
    for j in range(i+1, n):
        if bubble_selection_array[j] < bubble_selection_array[min_index]:
            min_index = j   
    bubble_selection_array[i], bubble_selection_array[min_index] = bubble_selection_array[min_index], bubble_selection_array[i]

print("Sorted array:", bubble_selection_array)
```
